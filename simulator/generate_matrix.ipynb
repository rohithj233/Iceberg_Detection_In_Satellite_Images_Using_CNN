```python
def set_indices_to_zero(matrix, indices):
    for index in indices:
        if isinstance(index, tuple):
            matrix[index[0]][index[1]] = 0
        elif isinstance(index, list):
            for i in range(index[0][0], index[1][0] + 1):
                for j in range(index[0][1], index[1][1] + 1):
                    matrix[i][j] = 0
```

```python
set_indices_to_zero(matrix, indices)
```
